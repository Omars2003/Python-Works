```python
import sympy as sp

# Definición de la función y su derivada
x = sp.symbols('x')
equation = input("Ingresa la ecuación f(x): ")
equation = sp.sympify(equation)  # Convierte la entrada del usuario en una expresión SymPy
f = sp.lambdify(x, equation, 'numpy')
f_prime = sp.diff(equation, x)
f_prime = sp.lambdify(x, f_prime, 'numpy')

# Valor inicial
x0 = float(input("Ingresa el valor inicial (x0): "))

# Tolerancia y máximo de iteraciones
tolerance = 1e-6
max_iterations = 100

# Algoritmo de Newton-Raphson
for i in range(max_iterations):
    f_x0 = f(x0)
    f_prime_x0 = f_prime(x0)
    if abs(f_prime_x0) < 1e-8:
        print("La derivada es cercana a cero. No se puede continuar.")
        break

    x1 = x0 - f_x0 / f_prime_x0
    if abs(x1 - x0) < tolerance:
        print(f"La raíz aproximada es: {x1} (con {i + 1} iteraciones)")
        break

    x0 = x1

if i == max_iterations - 1:
    print("El método de Newton-Raphson no convergió después de 100 iteraciones.")
```
